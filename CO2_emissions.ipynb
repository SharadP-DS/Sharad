```python
# Industries wise emissions in India
f[f['Country'] == 'India']
```

```python
# India's CO2 emissions in each sector
f.query('Country=="India"')
```

```python
f.query('Country=="India"')
```

```python
# North America CO2 emissions above a specific 140 tonnes criteria
f[(f['Region'] == 'North America') & (f["Co2_Emissions_MetricTons"]>140)]
```

```python
# average of emissions from various sectors
f.loc[:,['Co2_Emissions_MetricTons','Energy_Consumption_TWh','Automobile_Co2_Emissions_MetricTons','Industrial_Co2_Emissions_MetricTons','Agriculture_Co2_Emissions_MetricTons']].mean()
```

```python
f[:2]
```

```python
# Display average Carbon Dioxide emissions in metric tonnes w.r.t country
f.groupby('Region')['Co2_Emissions_MetricTons'].mean().round(2)  # sorts data A-Z
```

```python
f
```

```python
# Region wise emission analysis
f.groupby('Region')[['Energy_Consumption_TWh','Co2_Emissions_MetricTons','Population_Millions','GDP_Billion_USD','Urbanization_Percentage']].mean().round(2)
```

```python
# Country wise emission levels
f.groupby('Country')[['Energy_Consumption_TWh','Co2_Emissions_MetricTons','Population_Millions','GDP_Billion_USD','Urbanization_Percentage']].mean().round(2)
```

```python
# Sector wise highly emitting industries
f.groupby('Industry_Type')[['Energy_Consumption_TWh','Co2_Emissions_MetricTons']].mean().round(2)
```

```python
# Year and country wise emissions and rise in them
f.groupby(['Year','Country'])[['Energy_Consumption_TWh','Co2_Emissions_MetricTons']].mean().round(2)
```

```python
# Year wise increase in emission , population levels and rise in GDP and urbanisation
f.groupby('Year')[['Energy_Consumption_TWh','Co2_Emissions_MetricTons','Population_Millions','GDP_Billion_USD','Urbanization_Percentage']].mean().round(2)
```
